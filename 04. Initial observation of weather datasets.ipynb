```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stations = ['Karpos', 'Centar', 'Lisice', 'Rektorat', 'Miladinovci']
```

### Plot present vs. missing values

```python
for station in stations:
    df = pd.read_csv(f'./data/formatted-weather/{station}', index_col=0)
    plt.figure(figsize=(17, 8))
    sns.heatmap(df.isnull(), cbar=False).set_title(f'{station}', fontsize=20)
    plt.show()
```
